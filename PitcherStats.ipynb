```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
```

```python
pitch_data = 'Resources/pitch_data.csv'
pitch_data_df = pd.read_csv(pitch_data)
pitch_data_df.head()
```

```python
# data types
pitch_data_df.dtypes
```

```python
# number of null values in each column
pitch_data_df.isnull().sum()
```

```python
# create a new dataframe containgin only player_a stats
a_df = pitch_data_df[pitch_data_df['pitcher_id'] == 'player_a']
a_df.head()
```

```python
# create a new dataframe containgin only player_b stats
b_df = pitch_data_df[pitch_data_df['pitcher_id'] == 'player_b']
b_df.head()
```

```python
# get sum of total pitches by pitch type for player_a
a_pitch_type_count = a_df.groupby(["pitch_type"]).count()["pitch_id"]
a_pitch_type_count
```

```python
# get the sum of all the pitches for player_a
a_total_pitches = a_df["pitch_id"].count()
a_total_pitches
```

```python
# Calculate the percentage of pitches for each pitch type for player_a
a_pitch_type_percents = 100 * a_pitch_type_count / a_total_pitches
a_pitch_type_percents
```

```python
# Import mpl to change the plot configurations using rcParams.
import matplotlib as mpl

# Build Pie Chart
plt.subplots(figsize=(15, 9))
plt.pie(a_pitch_type_percents,
    labels=["changeup", "curveball", "cutter", "four_seam", "sinker", "slider"],
    colors=['lightgreen', 'lightcoral', 'lightskyblue', 'lavender', 'lightsalmon', 'lightblue'],
    explode=[0, 0, 0, 0.06, 0, 0],
    autopct='%1.1f%%')
plt.title("Player A % of Total Pitches by Pitch Type")

# Change the default font size from 10 to 12.
mpl.rcParams['font.size'] = 12

# Save Figure
plt.savefig("Analysis/PlayerA_PitchCount.png")

# Show Figure
plt.show()
```
